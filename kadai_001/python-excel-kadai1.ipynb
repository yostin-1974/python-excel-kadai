```python
from openpyxl import Workbook
from datetime import datetime

# ブック作成
wb = Workbook()
worksheet = wb.active

# タイトル
worksheet["B2"] = "請求書"

# 会社情報
worksheet["B4"] = "株式会社ABC"
worksheet["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
worksheet["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
worksheet["B7"] = "担当者名: 鈴木一郎"

# 請求書番号
worksheet["F4"] = "No."
worksheet["G4"] = "0001"

# 日付
today = datetime.today()
worksheet["F5"] = "日付"
worksheet["G5"] = today.strftime("%Y/%m/%d")

# 商品情報
header_row = 10
headers = ["商品名", "数量", "単価", "金額"]

for col_num, header in enumerate(headers, start=2):
    cell = worksheet.cell(row=header_row, column=col_num)
    cell.value = header

# データ
items = [
    ["商品A", 2, 10000],
    ["商品B", 1, 15000]
]

start_row = 11
total = 0

for row_idx, item in enumerate(items, start=start_row):
    print(f"Processing row {row_idx} with item: {item}")  # デバッグ用の出力
    worksheet.cell(row=row_idx, column=2).value = item[0]  # 商品名
    worksheet.cell(row=row_idx, column=3).value = item[1]  # 数量
    worksheet.cell(row=row_idx, column=4).value = item[2]  # 単価
    count = item[1]
    price = item[2]
    amount = count * price
    worksheet.cell(row=row_idx, column=5).value = amount  # 金額
    total += amount

start_row = worksheet.max_row + 1  # データの最終行の次の行を取得
worksheet.cell(row=start_row, column=5).value = total  # 合計金額を入力

# 合計表示
worksheet.cell(row=start_row + 2, column=2).value = "合計"
worksheet.cell(row=start_row + 2, column=5).value = total

# 消費税表示
worksheet.cell(row=start_row + 3, column=2).value = "消費税"
tax = int(total / 10)  # 消費税は合計金額の10%
worksheet.cell(row=start_row + 3, column=5).value = tax

# 税込み合計表示
worksheet.cell(row=start_row + 4, column=2).value = "税込合計"
worksheet.cell(row=start_row + 4, column=5).value = total + tax

# ファイル保存
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"
wb.save(filename)
```
